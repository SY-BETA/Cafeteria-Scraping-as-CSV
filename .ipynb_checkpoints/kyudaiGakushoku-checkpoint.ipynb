```python
import requests

r = requests.get('http://www.coop.kyushu-u.ac.jp/shokudou/month_menu.html')
# r = requests.get('https://www.crummy.com/software/BeautifulSoup/bs4/doc/')
# r.encoding = 'UTF-8'
"""shift_fisに注意"""
r.encoding = 'shift_jis'
r.encoding
```

```python
r.text
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
text = soup.find('tr').get_text()
rows = soup.find('tr')
rows
```

```python
import csv

with open("datas/td_data.csv", "w", encoding="utf-8") as file:
    writer = csv.writer(file)
    for row in rows:
        csvRow = []
        for cell in row.findAll(['td', 'th']):
            csvRow.append(cell.get_text())
        writer.writerow(csvRow)
        print("Saved!!")
```
